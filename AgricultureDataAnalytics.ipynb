# 3. Hypothesis Testing (T-Test, ANOVA)
Purpose: Check if means differ significantly (e.g., yield across seasons or soil types).

T-Test
Use when: Comparing means of two groups (e.g., yield in spring vs summer)

# Example: T-Test in Python

```python
import scipy.stats as stats
import pandas as pd

# Simulated yield data for two seasons
spring_yield = [30, 28, 35, 33, 31]
summer_yield = [25, 22, 27, 20, 23]

# Independent T-Test
t_stat, p_value = stats.ttest_ind(spring_yield, summer_yield)

print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print(" Significant difference in yield between Spring and Summer.")
else:
    print(" No significant difference in yield.")
```

# 3a ANOVA (Analysis of Variance)
Use when: Comparing means of 3 or more groups (e.g., yield on clay, sand, loam soils)

# Example: ANOVA in Python

```python
# Simulated yield data for different soil types
clay_yield = [20, 21, 19, 22]
sand_yield = [25, 26, 24, 27]
loam_yield = [30, 32, 31, 29]

# One-Way ANOVA
f_stat, p_value = stats.f_oneway(clay_yield, sand_yield, loam_yield)

print(f"F-statistic: {f_stat:.2f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print(" Significant difference in yield across soil types.")
else:
    print(" No significant difference in yield across soils.")
```

# 4. Linear & Multiple Regression
Purpose: Predict Yield based on input features.

# 4.1 Linear Regression with Python
Use Case: Predicting a dependent variable y from a single independent variable x.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
area = np.array([500, 1000, 1500, 2000, 2500]).reshape(-1, 1)
price = np.array([100000, 200000, 300000, 400000, 500000])

# Create and train model
model = LinearRegression()
model.fit(area, price)

# Predict
predicted_price = model.predict([[1800]])
print(f"Predicted price for 1800 sq.ft: {predicted_price[0]:,.0f}")

# Plot
plt.scatter(area, price, color='blue')
plt.plot(area, model.predict(area), color='red')
plt.xlabel('Area (sq.ft)')
plt.ylabel('Price (INR)')
plt.title('Linear Regression')
plt.show()
```

# 4.2 Multiple Linear Regression
Use Case: Predicting y using multiple independent variables like area, number of rooms, and location score.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Sample dataset
data = {
    'Area': [1000, 1500, 2000, 2500, 3000],
    'Rooms': [2, 3, 4, 4, 5],
    'Location_Score': [3, 4, 5, 4, 5],
    'Price': [150000, 250000, 350000, 400000, 500000]
}

df = pd.DataFrame(data)

# Independent variables (X) and dependent variable (y)
X = df[['Area', 'Rooms', 'Location_Score']]
y = df['Price']

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Prediction
test_input = pd.DataFrame([[2200, 3, 4]], columns=['Area', 'Rooms', 'Location_Score'])
predicted = model.predict(test_input)
print(f"Predicted price: {predicted[0]:,.0f}")
```
