```python
import matplotlib.pyplot as plt
```

```python
x = [1,3,5,7,9,10]
y = [2,4,6,8,12,17]
```

```python
plt.plot(x,y)
plt.title("tampilan line")
plt.xlabel("sumbu x")
plt.ylabel("sumbu y")
plt.show()
```

```python
import matplotlib.pyplot as plt
Angka = ["a","b","c","d","e"]
Nilai = [12,23,3,5,8]
plt.bar(Angka,Nilai)
plt.title("membuat bar")
plt.xlabel("angka")
plt.ylabel("Nilai")
plt.show()
```

```python
kategori= ["gambar a", "gambar b", "gambar c"]
ukuran = [ 20,30,50]
plt.pie(ukuran , labels= kategori, autopct ='%1.0f%%')
plt.title(" pie chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)
plt.hist(data,bins= 30, edgecolor= "black")
plt.title(" latihan histogram")
plt.xlabel("angka")
plt.ylabel("frekuensi kemunculan ")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randn(50)
y = 2*x +1+0.1*np.random.randn(50)
plt.scatter(x,y)
plt.title("latihan scatter")
plt.xlabel(" sumbu x")
plt.ylabel(" sumbu y")
plt.show()
```
